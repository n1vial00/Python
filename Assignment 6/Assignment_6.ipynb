```python
#Assignment 6

import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import numpy as np

projectId = "third-serenity-367111"
SQL = """
SELECT * FROM `bigquery-public-data.world_bank_global_population.population_by_country` 
"""
i_df = pd.read_gbq(SQL, projectId)
```

```python
df_tr = i_df.T #Transpose

df_tr.columns = df_tr.iloc[0]

df_tr = df_tr.drop(df_tr.index[[0,1]])

df_tr.reset_index(inplace=True)
df_tr.columns.name = None
df_tr = df_tr.rename(columns = {'index':'year'})
df_tr['year'] = pd.to_numeric(df_tr['year'].str[5:9])
df_tr.drop(df_tr.tail(1).index,inplace=True)


#Figure 
plt.figure(num=None, figsize=(25, 10), dpi=300, facecolor='w', edgecolor='k')

plt.plot(df_tr['year'], df_tr['Italy'] / 1000000,  linewidth = '2')
plt.plot(df_tr['year'], df_tr['Spain'] / 1000000, linewidth = '2')
plt.plot(df_tr['year'], df_tr['France'] / 1000000, linewidth = '2')
plt.title('Population in Italy, Spain and France since 1960')
plt.ylabel('Million people')
plt.legend(['Italy', 'Spain', 'France'])
plt.grid(True)

plt.show()


#Stacked Bar Plot 
plt.figure(num=None, figsize=(25, 10), dpi=300, facecolor='w', edgecolor='k')

plt.bar(df_tr['year'], df_tr['Sweden'] / 1000000, color = '#FECD00')
plt.bar(df_tr['year'], df_tr['Denmark'] / 1000000, color = '#C00D2E')
plt.bar(df_tr['year'], df_tr['Finland'] / 1000000, color = '#002E69')
plt.bar(df_tr['year'], df_tr['Norway'] / 1000000, color = '#000000') # Because oil
plt.bar(df_tr['year'], df_tr['Iceland'] / 1000000, color = '#D51D33')
plt.title('Population in Nordic countries since 1960')
plt.ylabel('Million people')
plt.legend(['Sweden', 'Denmark', 'Finland', 'Norway', 'Iceland'])
plt.grid(True)

plt.show()
```
