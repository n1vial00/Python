```python
import matplotlib.pyplot as plt
import plotly.express as px
data = px.data.gapminder().query("continent=='Europe'")
data_FI = data[data['country'] == "Finland"]
data_SWE = data[data['country'] == "Sweden"]
x = data_FI['year']
y1 = data_FI['gdpPercap']
y2 = data_SWE['gdpPercap']
```

```python
#Line plot

plt.figure(num=None, figsize=(10, 5), dpi=200, facecolor='w', edgecolor='k')
#plt.plot(x,y1,linestyle = 'dotted', color = 'r') #Finland
plt.plot(x,y1) #Finland
#plt.plot(x,y2,linestyle = 'dashed', c = '#4CAF50', linewidth = '5') #Sweden
plt.plot(x,y2) #Sweden
plt.xlabel('Time')
plt.ylabel('GDP per capita')
plt.title('GDP per capita in Finland and Sweden')
plt.legend(['Finland', 'Sweden'])
plt.grid(True)
#plt.xlim(1960,1990)
#plt.ylim(5000,25000)
plt.show()
```

```python
#Scatter plot
plt.figure(num=None, figsize=(10, 5), dpi=200, facecolor='w', edgecolor='k')
plt.scatter(x,y1) #Finland
#plt.plot(x,y1) #Finland
plt.scatter(x,y2) #Sweden
#plt.plot(x,y2) #Sweden
plt.xlabel('Time')
plt.ylabel('GDP per capita')
plt.title('GDP per capita in Finland and Sweden')
plt.legend(['Finland', 'Sweden'])
plt.grid(True)
plt.xlim(1960,1990)
plt.ylim(5000,25000)
plt.show()
```

```python
#Scatter plot between same quantities
plt.figure(num=None, figsize=(5, 5), dpi=100, facecolor='w', edgecolor='k')
plt.scatter(y1,y2) 
plt.xlabel('Time')
plt.ylabel('GDP per capita')
plt.title('GDP  in Finland vs. GDP in Sweden')
plt.grid(True)
plt.show()
```

```python
#Scatter plot between same quantities, color indicates third parameter
plt.figure(num=None, figsize=(5, 5), dpi=100, facecolor='w', edgecolor='k')
plt.scatter(y1,y2,c=x)
plt.xlabel('Time')
plt.ylabel('GDP per capita')
plt.title('GDP  in Finland vs. GDP in Sweden')
plt.grid(True)
plt.colorbar()
plt.axis("square")
plt.show()

```

```python
#Dummy errors
import numpy as np
random_data = np.random.rand(x.shape[0])*10000

#Error bar plot 
plt.figure(num=None, figsize=(10, 5), dpi=100, facecolor='w', edgecolor='k')
plt.errorbar(x,y1,yerr = random_data, fmt='o', linewidth=2, capsize=6)
plt.xlabel('Time')
plt.ylabel('GDP per capita')
plt.title('GDP per capita in Finland ')
plt.grid(True)
plt.show()
```

```python
#Bar plot
plt.figure(num=None, figsize=(10, 5), dpi=100, facecolor='w', edgecolor='k')
width = 2
plt.bar(x- width/2,y1,width = width ) #Finland
plt.bar(x+ width/2,y2,width = width) #Sweden
plt.xlabel('Time')
plt.ylabel('GDP per capita')
plt.title('GDP per capita in Finland and Sweden')
plt.legend(['Finland', 'Sweden'])
plt.grid(True)
plt.xticks(x)
plt.show()
```

```python
#Stacked bar plot
plt.figure(num=None, figsize=(10, 5), dpi=100, facecolor='w', edgecolor='k')
plt.bar(x,y1,width = 3) #Finland
plt.bar(x,y2,bottom = y1, width = 3) #Sweden
plt.xlabel('Time')
plt.ylabel('GDP per capita')
plt.title('GDP per capita in Finland and Sweden')
plt.legend(['Finland', 'Sweden'])
plt.grid(True)

# plt.show()
```

```python
data = px.data.gapminder()
data_2007 = data[data['year'] == 2007]
GDP =data_2007['gdpPercap']
```

```python
#Histogram
fig, ax = plt.subplots(num=None, figsize=(10, 5), dpi=100, facecolor='w', edgecolor='k')
plt.hist(GDP,bins = np.arange(0,50000,1000))
ax.set_yticks(np.arange(0,22,5), minor=False)
plt.ylabel('Number of countries')
plt.xlabel('GDP per capita')
plt.title('Histogram of GDP per capita in Europe in 2007')
plt.grid(True)
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Cat', 'Dog', 'Hamster', 'Goldfish'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

plt.pie(sizes, labels=labels, shadow=True, autopct='%1.1f%%', startangle=90, explode=explode)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pets')
plt.show()
```
